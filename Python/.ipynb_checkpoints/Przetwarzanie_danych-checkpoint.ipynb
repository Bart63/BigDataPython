```python
import pandas as pd
```

```python
athlete_events_fname = 'athlete_events.csv'
athlete_events_data_dict_fname = 'athlete_events_data_dictionary.csv'
```

Opis kolumn:

```python
with open(athlete_events_data_dict_fname, 'r') as f:
    for line in f.readlines():
        print(line)
```

```python
athlete_df = pd.read_csv(athlete_events_fname, index_col=0)
athlete_df
```

# Zadanie 1

## Przygotować zestawienie przedstawiające łączną liczbę zawodników na olimpiadach letnich i zimowych w kolejnych latach.

Interesują nas dwie kolumny: Year i Season

Osoba (Name), która uczestniczyła podczas danej olimpiady (Year, Season) w wielu dyscyplinach, wydarzeniach (Sport, Event) jest liczona wiele razy - jako osobny zawodnik

```python
athletes_by_season_df = pd.DataFrame(athlete_df.groupby(['Year', 'Season']).size(), columns=['AthletesCount'])
athletes_by_season_df
```

# Zadanie 2

## Przygotować zestawienie przedstawiające średnią liczbę startujących kobiet i mężczyzn na olimpiadach letnich dla każdego narodowego komitetu olimpijskiego.

Chcemy na początku otrzymać olimpiady letnie. Uzyskujemy następnie liczbę olimpiad letnich jakie miały miejsce. Potem dzielimy liczbę zawodników podzielonych na komitet i płeć przez liczbę olimpiad

```python
only_summer_games_df = athlete_df[athlete_df['Season'] == 'Summer']
only_summer_games_df
```

```python
unique_games = only_summer_games_df['Games'].nunique()
unique_games
```

```python
avg_sex_by_NOC = pd.DataFrame(
    only_summer_games_df.groupby(['NOC', 'Sex']).size() / unique_games, 
    columns=['AvgAthletesPerGame']
)
avg_sex_by_NOC
```

# Zadanie 3

## Przygotować zestawienie przedstawiające 5 narodowych komitetów olimpijskich, które wystawiły łącznie najwięcej różnych zawodników oraz liczbę ich zawodników uzależnioną od liczby olimpiad, w których dany zawodnik brał udział.

Liczba zawodników danego komitetu jest uzależniona od liczby olimpiad, w których dany zawodnik brał udział (Games). 

Przykład: Zawodnik o imieniu A był na 4 różnych olimpiadach (Games) i na każdym brał udział w różnych dyscyplinach i eventach (Sport, Event). Do sumy liczy się jako 4

```python
five_biggest_df = pd.DataFrame(athlete_df.groupby(['NOC', 'Games', 'Name']).size().groupby(['NOC']).count().nlargest(5), columns=['AthletesCount'])
five_biggest_df
```
