```python
import pandas as pd
import time
```

```python
salaries_df = pd.read_csv('Salaries.csv', header=0)
salaries_df
```

```python
batting_df = pd.read_csv('Batting.csv', header=0)
batting_df
```

# 1. identyfikatory wszystkich drużyn (teamID), w których w roku 1985 każdy gracz zarabiał powyżej 100000.

```python
start_time = time.time()

salaries_min_1985_df = salaries_df[salaries_df['yearID']==1985].groupby('teamID').min()
teamID_list = salaries_min_1985_df[salaries_min_1985_df['salary']>100000].index.tolist()

end_time = time.time()

print('Rozwiązanie:', teamID_list)
print(f'Operacja zajęła {end_time-start_time} sekund')
```

# 2. identyfikatory wszystkich drużyn (teamID) wraz ze średnim wynagrodzeniem w tych drużynach w 1998.

```python
start_time = time.time()

salaries_avg_1998_df = salaries_df[salaries_df['yearID']==1998].groupby('teamID').mean()['salary']

end_time = time.time()

print('Rozwiązanie:', salaries_avg_1998_df)
print(f'Operacja zajęła {end_time-start_time} sekund')
```

# 3. dla każdej ligii (lgID) liczbę drużyn w 1999.

```python
start_time = time.time()

teams = salaries_df[salaries_df['yearID']==1999].groupby('lgID').nunique()['teamID']

end_time = time.time()

print('Rozwiązanie:', teams)
print(f'Operacja zajęła {end_time-start_time} sekund')
```

# 4. top 10 graczy z największą liczbą uderzeń (H - hits) w 1988. Wypisz playerID i liczbę tych uderzeń.

```python
start_time = time.time()

top_10_player_hits = batting_df[batting_df['yearID']==1988].nlargest(10, 'H')[['playerID', 'H']]

end_time = time.time()

print('Rozwiązanie:\n', top_10_player_hits)
print(f'Operacja zajęła {end_time-start_time} sekund')
```

# 5. graczy, którzy rozgrywali najwięcej gier w 1980 (G_batting - kolumna nr 7). Wypisz playerID i liczbę tych gier.

```python
start_time = time.time()

top_10_player_games = batting_df[batting_df['yearID']==1980].nlargest(10, 'G_batting')[['playerID', 'G_batting']]

end_time = time.time()

print('Rozwiązanie:\n', top_10_player_games)
print(f'Operacja zajęła {end_time-start_time} sekund')
```

# 6. graczy, którzy w drużynie ML1 wykonali najwięcej biegów (runs - R). Wypisz playerID i liczbę R. 

```python
start_time = time.time()

top_10_player_runs = batting_df[batting_df['teamID']=='ML1'].nlargest(10, 'R')[['playerID', 'R']]

end_time = time.time()

print('Rozwiązanie:\n', top_10_player_runs)
print(f'Operacja zajęła {end_time-start_time} sekund')
```

# 7. graczy, którzy zarobili w 2001 powyżej 500000 i mają więcej niż 50 biegów do bazy (homeruns - HR).

```python
start_time = time.time()

players_salary = set(salaries_df[(salaries_df['yearID']==2001) & (salaries_df['salary']>500000)]['playerID'].unique())
players_batting = set(batting_df[batting_df['HR']>50]['playerID'].unique())
selected = players_salary & players_batting

end_time = time.time()

print('Rozwiązanie:\n', selected)
print(f'Operacja zajęła {end_time-start_time} sekund')
```

# Wnioski

- Kod napisany w języku Python wykonywany jest znacznie szybciej niż polecenia w PigLatin.
- PigLatin jest kompilowany do zadań MapReduce.
- W PigLatin średniki na końcu poleceń są obowiązkowe.
- Dane w Pythonie zostały umieszczone w pamięci operacyjnej, w PigLatin dane były zczytywane z pamięci masowej.
