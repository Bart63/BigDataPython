```python
import pandas as pd
import time
```

```python
salaries_df = pd.read_csv('Salaries.csv', header=0)
salaries_df
```

```python
batting_df = pd.read_csv('Batting.csv', header=0)
batting_df
```

# 1. identyfikatory wszystkich druÅ¼yn (teamID), w ktÃ³rych w roku 1985 kaÅ¼dy gracz zarabiaÅ‚ powyÅ¼ej 100000.

```python
start_time = time.time()

salaries_min_1985_df = salaries_df[salaries_df['yearID']==1985].groupby('teamID').min()
teamID_list = salaries_min_1985_df[salaries_min_1985_df['salary']>100000].index.tolist()

end_time = time.time()

print('RozwiÄ…zanie:', teamID_list)
print(f'Operacja zajÄ™Å‚a {end_time-start_time} sekund')
```

# 2. identyfikatory wszystkich druÅ¼yn (teamID) wraz ze Å›rednim wynagrodzeniem w tych druÅ¼ynach w 1998.

```python
start_time = time.time()

salaries_avg_1998_df = salaries_df[salaries_df['yearID']==1998].groupby('teamID').mean()['salary']

end_time = time.time()

print('RozwiÄ…zanie:', salaries_avg_1998_df)
print(f'Operacja zajÄ™Å‚a {end_time-start_time} sekund')
```

# 3. dla kaÅ¼dej ligii (lgID) liczbÄ™ druÅ¼yn w 1999.

```python
start_time = time.time()

teams = salaries_df[salaries_df['yearID']==1999].groupby('lgID').nunique()['teamID']

end_time = time.time()

print('RozwiÄ…zanie:', teams)
print(f'Operacja zajÄ™Å‚a {end_time-start_time} sekund')
```

# 4. top 10 graczy z najwiÄ™kszÄ… liczbÄ… uderzeÅ„ (H - hits) w 1988. Wypisz playerID i liczbÄ™ tych uderzeÅ„.

```python
start_time = time.time()

top_10_player_hits = batting_df[batting_df['yearID']==1988].nlargest(10, 'H')[['playerID', 'H']]

end_time = time.time()

print('RozwiÄ…zanie:\n', top_10_player_hits)
print(f'Operacja zajÄ™Å‚a {end_time-start_time} sekund')
```

# 5. graczy, ktÃ³rzy rozgrywali najwiÄ™cej gier w 1980 (G_batting - kolumna nr 7). Wypisz playerID i liczbÄ™ tych gier.

```python
start_time = time.time()

top_10_player_games = batting_df[batting_df['yearID']==1980].nlargest(10, 'G_batting')[['playerID', 'G_batting']]

end_time = time.time()

print('RozwiÄ…zanie:\n', top_10_player_games)
print(f'Operacja zajÄ™Å‚a {end_time-start_time} sekund')
```

# 6. graczy, ktÃ³rzy w druÅ¼ynie ML1 wykonali najwiÄ™cej biegÃ³w (runs - R). Wypisz playerID i liczbÄ™ R. 

```python
start_time = time.time()

top_10_player_runs = batting_df[batting_df['teamID']=='ML1'].nlargest(10, 'R')[['playerID', 'R']]

end_time = time.time()

print('RozwiÄ…zanie:\n', top_10_player_runs)
print(f'Operacja zajÄ™Å‚a {end_time-start_time} sekund')
```

# 7. graczy, ktÃ³rzy zarobili w 2001 powyÅ¼ej 500000 i majÄ… wiÄ™cej niÅ¼ 50 biegÃ³w do bazy (homeruns - HR).

```python
start_time = time.time()

players_salary = set(salaries_df[(salaries_df['yearID']==2001) & (salaries_df['salary']>500000)]['playerID'].unique())
players_batting = set(batting_df[batting_df['HR']>50]['playerID'].unique())
selected = players_salary & players_batting

end_time = time.time()

print('RozwiÄ…zanie:\n', selected)
print(f'Operacja zajÄ™Å‚a {end_time-start_time} sekund')
```
